```python
import numpy as np
from random import random
```

Utiizaremos las siguientes formulas paraproximar el valor de la integral:

  $\int_a^b g(x) dx=\int_a^b w(x)f(x)dx=E[g(x)]=\frac{1}{N} \sum_{i=1}^N g(x_i)  $  

$w(x)=g(x)b-a$ \\

$f(x)=\frac {1}{b-a}$

```python

def funcion(x):
  return np.sqrt(np.atan((x)))

def integral(a,b,N):
  I=0
  for i in range(N):
    g=funcion(random()*(b-a)+a)
    I=I+ g
  return I/N
```

```python
x=[]
for i in range(100):
  x.append(integral(0,1,100000))
integral(0,1,100000)
```

```python
from statistics import mean, stdev
import matplotlib.pyplot as plt

x0=0.62982384
plt.plot(x)
plt.xlabel('n√πmero de iteraciones')
plt.ylabel('valor de la integral')
plt.axhline(y=x0,color='r',linestyle='-')
plt.axhline(y=mean(x)+ stdev(x),color='r',linestyle= '--')
plt.axhline(y=mean(x)- stdev(x),color='r',linestyle= '--')
plt.title('Integral')
plt.grid(True)
plt.show()
```
