Código que genera una distribución normal a partir del método de las 12 uniformes. Los datos de la distribución son datos aleatorios generados por el mismo código. Al final graficamos en un histograma los datos simulados, con la media y desviacion estandar respecto a esos datos.

Sea x una variable aleatoria continua que sigue una distribución normal con parámetros $\mu$ y $\sigma$, se define su función de distribución de probabilidad como:
$$f(x)=\frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{1}{2}(\frac{x-μ}{\sigma})^2}$$ donde $-∞<x<∞$ \\

Para generar una distribución normal usamos el teorema del límite central.

Generamos n números aleatorios y consideramos
$$z=\frac{Y-\sum_{i+1}^n a_i \mu_i}{\sqrt{\sum_{i=1}^n} \sigma_i^2}$$
donde $Y=R_1+R_2+...+ R_n$

$$\therefore Z=\frac{\sum_{i=1}^n R_i - \frac{n}{2}}{\sqrt{\frac{n}{12}}}$$

$$\therefore Z=\sum_{i=1}^n R_i - 6 $$

$$x=\mu + \sigma(\sum_{i=1}^{12} R_i-6)$$

```python
from random import random
import matplotlib.pyplot as plt
from statistics import mean, stdev
```



La esencia de este código radica que en esta expresión:

$$x=\mu + \sigma(\sum_{i=1}^{12} R_i-6)$$

Función que realiza el método de las 12 uniformes

```python
def doce(media, desvi):
  R=random()
  for i in range(12):#ciclo que realiza la sumatoria de Ri terminos donde i=1...12
    R=R+random()
  R=R-6
  x=media + desvi*R #expresión para obtener los valores que siguen una dist. normal
  return x
```

Simulación de los datos de la función previa

```python
x=[]
media=0
desvi=1
for i in range(100000):
  x.append(doce(media,desvi)) #asignamos valores a la media y la desviación estándar
```

Graficamos los datos

```python
plt.hist(x, bins=500)
plt.axvline(x=mean(x), color = 'r', label=f"media={mean(x)}")
plt.axvline(x=mean(x)+stdev(x), color = 'r', linestyle='--', label=f"desviación estándar={stdev(x)}")
plt.axvline(x=mean(x)-stdev(x), color = 'r', linestyle='--')
plt.title('Simulación del método de las 12 uniformes')
plt.xlabel('x')
plt.ylabel('Valores')
plt.legend(loc='best')
plt.show()
```

```python
mediasimu = mean(x)
stdevsimu = stdev(x)

print("Media teórica:", media)
print("Media simulada:", mediasimu)

print("\nVarianza teórica:", desvi)
print("Varianza simulada:", stdevsimu)
```
