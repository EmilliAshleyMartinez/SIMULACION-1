Código que estima el valor de la integral definida de una función por el método MONTE CARLO.Cada parte importante del código se encuentra comentada en su respectiva linea.

```python
import numpy as np
from random import random
import matplotlib.pyplot as plt
from statistics import mean, stdev
import matplotlib.pyplot as plt
```

```python
#definimos la función
def funcion(x):
    return np.sqrt(np.arctan(x))
```

Para este metodo utilice las siguientes formulas clave

**generador de puntos**: \\
$x_i=a + [random 1 * (b-a)]$ \\
$y_i=random 2 *M $ \\
 donde M:cota \\
**puntos de éxito:**
$$y_i \leq f(x_i)$$

 y estimación de la **integral**:
 $$\int_a^b f(x)dx= \frac{exitos}{N}*(b-a)*M$$

```python
#definimos puntos
def puntos(a,b,M,N):
  exito=0
  total=N
#en este ciclo generamos los puntos de acuerdo a las formulas y clasificamos si
#es un punto de éxito no, es decir que esta en el área que nos interesa
  for i in range(N):
    xi=a + (random()*(b-a))
    yi=random()*M
    if yi <= funcion(xi):
      exito= exito + 1
  return (exito/N)*M*(b-a)
```

```python
valor=puntos(0,1,1,100000)
#Llamamos a la función y mostramos el resultado obtenido
print("Valor calculado para la integral:", valor)
```

```python
from inspect import Parameter
#este bloque es exclusivo para generar los datos que necesitamos para graficar
x=[]
for i in range(1,100): #generamos 100 datos para graficar
  x.append(puntos(0,1,1,i))
error= stdev(x)
print ("Error:",error)
```

```python
x0=0.62982384 #valor real
plt.plot(x)#graficamos los puntos o datos del bloque anterior
plt.xlabel('número de iteraciones')
plt.ylabel('valor de la integral')
plt.axhline(y=x0,color='r',linestyle='-') #recta del valor real
plt.axhline(y=mean(x)+ stdev(x),color='r',linestyle= '--') #intervalos del error calculados a desviaciones estandar de la media
plt.axhline(y=mean(x)- stdev(x),color='r',linestyle= '--')
plt.title('Integral por Monte Carlo')
plt.grid(True)
plt.show()
```
