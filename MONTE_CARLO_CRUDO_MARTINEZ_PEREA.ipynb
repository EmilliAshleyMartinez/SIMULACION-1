Código que estima el valor de la integral definida de una función por el método MONTE CARLO CRUDO. Cada parte importante del código se encuentra comentada en su respectiva linea.

```python
#Importamos librerias requeridas para:
#generar números aleatorios
import numpy as np
from random import random
#grafica y error
from statistics import mean, stdev
import matplotlib.pyplot as plt
```

```python
#definimos la función
def funcion(x):
  return np.sqrt(np.atan((x)))
```

Utiizaremos las siguientes formulas para aproximar el valor de la integral:

  $\int_a^b g(x) dx=\int_a^b w(x)f(x)dx=E[g(x)]=\frac{1}{N} \sum_{i=1}^N g(x_i)  $  \\

$w(x)=g(x)b-a$ \\

$f(x)=\frac {1}{b-a}$ \\
A diferencia de montecarlo, no determinamos si los puntos son de exito o no porque trabajamos con el promedio de los puntos generados

```python
#función que genera la integral
def integral(a,b,N):
  I=0
  for i in range(N):
    g=funcion(random()*(b-a)+a) #evaluamos en los puntos; los puntos fueron generados con números aleatorios y delimitados al intervalo de integral
    I=I+ g #generamos la sumatoria de g
  return I/N #calculamos promedio
```

```python
#Llamamos a la función y mostramos el resultado obtenido
print("Valor calculado para la integral:", integral(0,1,100000))
```

```python
#este bloque es exclusivo para generar los datos que necesitamos para graficar generamos 100 datos
x=[]
for i in range(1,100):
  x.append(integral(0,1,i))

error= stdev(x)
print ("Error:",error)#este error es una estimación aproximada para los 1000 datos
```

```python
x0=0.62982384#valor real de la integral
plt.plot(x)#graficamos los datos generados
plt.xlabel('nùmero de iteraciones')
plt.ylabel('valor de la integral')
plt.axhline(y=x0,color='r',linestyle='-') #recta del valor real
#intervalos del error calculados a desviaciones estandar de la media
plt.axhline(y=mean(x)+ stdev(x),color='r',linestyle= '--')
plt.axhline(y=mean(x)- stdev(x),color='r',linestyle= '--')
#mostramos la integral
plt.title('Integral arctan(x) por Monte Carlo Crudo')
plt.grid(True)
plt.show()
```
