**Generador congruencial de numeros aleatorios** \\
Este programa representa el funcionamiento de un generador congruencial. Como elemento principal se utiliza un ciclo *for* para poder iterar la formula ya establecida en la teoría y dentro de ese ciclo con un condicional se realizó el contador para calcular el ciclo de vida.
Cada parte del código se encuentra comentada en su respectiva linea.

```python
from tabulate import tabulate

def generador(seed, a, c, m, n):
#Inicialización de variables
  valores = [] #aqui se almacenan los xn del generador
  xn = seed #semilla o numero inicial
  tabla = [] #lista que permite almacenar los datos a tabular
  ciclo = 0 #inicio contador

  for i in range (n): #el ciclo for va de n=0 a n número que se desean generar
    xn=((a*xn)+c) % m  #formula del generador congruencial, en n=0 tenemos xn=semilla
    if xn in valores: #condicional que busca desde i hasta n, en la lista de valores el mismo xn
      ciclo = i #una vez encontrado el xn repetido ese indice será el valor de nuestro ciclo de vida
      break #para que cuando se encuentre no se siga repitiendo el if mientras corre el for

    valores.append(xn) #agrega los xn generados en el ciclo for a la lista de valores
    tabla.append([i, xn, xn/m]) #tabula los datos agregados a la lista tabla

#Imprime el valor obtenido del condicional
  print(f"El ciclo de vida del generador es: {ciclo}")
#Imprime tabla de resultados
  print(tabulate(tabla, headers=["n", "Xn", "Xn/m"], tablefmt="grid"))


#Ingreso de datos
seed = 7
a = 13
c = 9
m = 128
#cantidad de números aleatorios que se desean generar
n=200
#llamado a la función para ejecutar
generador(seed, a, c, m, n)
```
